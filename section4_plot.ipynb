```python
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
```

```python
x = np.linspace(0,1,100)
y = np.sin(2*np.pi*x)

plt.plot(x,y,color="blue",linestyle=":",linewidth=3,label="sin") # linstyle, linewidth,label
plt.legend()
```

```python
x = np.linspace(0,1,100)
y = np.sin(2*np.pi*x)
y1 = y/2.
plt.plot(x,y,color="red") # linstyle, linewidth,label
plt.plot(x,y1,color="blue")
```

```python
x = np.arange(0,10)
y = 1 / x
plt.bar(x,y,color='red',width=0.9)
```

```python
sst = xr.open_dataset("../DATA/HadISST_sst_2x2.nc")['sst']
sstm = sst.mean("time")
x = sst.lon
y = sst.lat
plt.contourf(x,y,sstm,
             cmap="RdBu_r",
             levels=np.linspace(1,40,20),
             extend="both")
plt.colorbar()
```

```python
u = xr.open_dataset("../DATA/ncep2_180x90/uwnd.10m.mon.mean.nc")['uwnd']
v = xr.open_dataset("../DATA/ncep2_180x90/vwnd.10m.mon.mean.nc")['vwnd']


def mean(data):
    return data.mean("time")[0]


um = mean(u)
vm = mean(v)
x1 = um.lon
y1 = um.lat
plt.quiver(x1,y1,um,vm)
```
