## 计算

```python
# 求平均

sst_ave = sst.mean("time")
sst_ave.plot.contourf(levels=np.linspace(1,40,20))
```

```python
sst_ave = sst.mean("lon")
sst_ave.plot.contourf(levels=np.linspace(1,40,20))
```

```python
sst_ave = sst.mean(("lon","lat"))
sst_ave.plot()
```
